Вопрос 2¶

Теперь вы распечатаете другое сообщение по вашему выбору. Для этого измените print("Ваше сообщение здесь!") на другое сообщение. Например, вы можете изменить его на что-то вроде:

print("Доброе утро!")
print("Я учусь программировать :D")
Или вы можете посмотреть, что произойдет, если вы напишете что-то вроде print("3+4"). Возвращает ли он 7 или просто воспринимает «3+4» как еще одно сообщение?

Убедитесь, что ваше сообщение заключено в кавычки ("), а само сообщение не содержит кавычек. Например, это приведет к ошибке: print("Она сказала "отличная работа" и дала мне пять! "), поскольку сообщение содержит кавычки. Если вы решите пройти курс Python после завершения этого курса, вы узнаете больше о том, как избежать этой ошибки, в Уроке 6.
