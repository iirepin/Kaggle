# Вступление
Добро пожаловать на промежуточный курс машинного обучения от Kaggle!

Если у вас есть некоторый опыт в машинном обучении и вы хотели бы узнать, как быстро улучшить качество ваших моделей, вы обратились по адресу! На этом курсе вы повысите свой опыт в машинном обучении, научившись::

разбирайтесь с типами данных, часто встречающимися в реальных наборах данных (пропущенные значения, категориальные переменные),
проектируйте конвейеры для повышения качества кода машинного обучения,
используйте передовые методы проверки моделей (перекрестная проверка).,
создавайте самые современные модели, которые широко используются для победы в конкурсах Kaggle (XGBoost), и
избегайте распространенных и важных ошибок в обработке данных (утечек).
Попутно вы будете применять свои знания, выполняя практические упражнения с использованием реальных данных по каждой новой теме. В практических упражнениях используются данные конкурса цен на жилье для пользователей Kaggle Learn, в котором вы будете использовать 79 различных независимых переменных (таких как тип крыши, количество спален и ванных комнат) для прогнозирования цен на жилье. Вы сможете оценить свой прогресс, отправляя прогнозы на это соревнование и наблюдая, как растет ваша позиция в таблице лидеров!

# Предпосылки¶
Вы готовы к этому курсу, если раньше создавали модель машинного обучения и знакомы с такими темами, как проверка модели, недостаточное и избыточное оснащение, а также случайные леса.

Если вы совершенно новичок в машинном обучении, пожалуйста, ознакомьтесь с нашим вводным курсом по машинному обучению, который содержит все необходимое для подготовки к этому курсу.

# ваш ход
Переходите к первому упражнению, чтобы узнать, как отправлять прогнозы на конкурс Kaggle, и определить, что вам, возможно, потребуется просмотреть перед началом работы.
