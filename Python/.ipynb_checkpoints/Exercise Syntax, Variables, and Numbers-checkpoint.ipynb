```python
############ Код настройки — не трогайте эту часть #######################
# Если вам интересно, вот примеры списков. Мы поговорим о
# них подробнее через несколько уроков. А пока просто знайте, что это
# еще один тип объекта Python, например int или float.
a = [1, 2, 3]
b = [3, 2, 1]

######################################################################

# Ваш код здесь. Поменяйте местами значения, на которые ссылаются a и b.
# Если вы застряли, вы всегда можете раскомментировать одну или обе строки в
# следующей ячейке, чтобы получить подсказку или взглянуть на решение.

a, b = b, a
print( a, "\n", b)
######################################################################
```

Самое простое решение — использовать третью переменную для временного хранения одного из старых значений. Например:

tmp = a
a = b
b = tmp
Если вы читали много кода Python, вы могли видеть следующий трюк для замены двух переменных в одной строке:

a, b = b, a

Алиса, Боб и Кэрол договорились объединить свои конфеты на Хэллоуин и разделить их поровну между собой. Ради их дружбы все оставшиеся конфеты будут разбиты. Например, если они вместе принесут домой 91 конфету, они возьмут по 30 и разобьют 1.

Запишите ниже арифметическое выражение, чтобы вычислить, сколько конфет им нужно разбить для заданного груза.

```python
# Переменные, представляющие количество конфет, собранных Алисой, Бобом и Кэрол
alice_candies = 121
bob_candies = 77
carol_candies = 109

# Ваш код здесь! Замените правую часть этого назначения выражением
# включающим alice_candies, bob_candies и carol_candies
to_smash = (alice_candies + bob_candies + carol_candies) % 3
print(to_smash)
```
