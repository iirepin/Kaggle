Давайте начнем с сильного удара молнии, чтобы разогреться. Какова длина строк, приведенных ниже?

Для каждой из пяти строк, приведенных ниже, спрогнозируйте, что вернет функция len() при передаче этой строки. Используйте переменную длину, чтобы записать свой ответ, затем запустите ячейку, чтобы проверить, были ли вы правы.

# 0a.

```python
a = ""
length = len(a)
print(length)
```

Правильный:

Длина пустой строки равна нулю. Обратите внимание, что пустая строка также является единственной строкой, которую Python считает ложной при преобразовании в логическое значение.

# 0b.

```python
b = "it's ok"
length = len(b)
print(length)
```

Правильный:

Имейте в виду, что Python использует пробелы (и знаки препинания) при подсчете длины строки.

# 0c.

```python
c = 'it\'s ok'
length = len(c)
print(length)
```

Правильный:

Несмотря на то, что мы используем другой синтаксис для ее создания, строка c идентична строке b. В частности, обратите внимание, что обратная косая черта не является частью строки, поэтому она не влияет на ее длину.

# 0d.

```python
d = """hey"""
length = len(d)
print(length)
```

Правильный:

Тот факт, что эта строка была создана с использованием синтаксиса тройных кавычек, не имеет никакого значения с точки зрения ее содержания или длины. Эта строка в точности такая же, как "hey".

# 0e.

```python
e = '\n'
length = len(e)
print(length)
```

Правильный:

Символ новой строки - это всего лишь один символ! (Хотя мы представляем его в Python, используя комбинацию из двух символов).
