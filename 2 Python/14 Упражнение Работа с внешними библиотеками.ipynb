 –ò–Ω–æ–≥–¥–∞ –∏–º—è –≥–æ–Ω—â–∏–∫–∞ –Ω–µ –±—ã–ª–æ –∑–∞–ø–∏—Å–∞–Ω–æ
    {'–Ω–∞–∑–≤–∞–Ω–∏–µ': –ù–µ—Ç, '–ø—Ä–µ–¥–º–µ—Ç—ã': ['–≥—Ä–∏–±—ã',], '–æ—Ç–¥–µ–ª–∫–∞': 2},
    {'–Ω–∞–∑–≤–∞–Ω–∏–µ': '–ñ–∞–±–∞', '–ø—Ä–µ–¥–º–µ—Ç—ã': ['–∑–µ–ª–µ–Ω–∞—è —Å–∫–æ—Ä–ª—É–ø–∞', '–≥—Ä–∏–±'], '–æ—Ç–¥–µ–ª–∫–∞': 1},

"–ø—Ä–µ–¥–º–µ—Ç—ã" - —ç—Ç–æ —Å–ø–∏—Å–æ–∫ –≤—Å–µ—Ö —É—Å–∏–ª–∏–≤–∞—é—â–∏—Ö –ø—Ä–µ–¥–º–µ—Ç–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ –≥–æ–Ω—â–∏–∫ —Å–æ–±—Ä–∞–ª –≤ —ç—Ç–æ–π –≥–æ–Ω–∫–µ, –∞ "—Ñ–∏–Ω–∏—à" - —ç—Ç–æ –∏—Ö —Ä–∞—Å–ø–æ–ª–æ–∂–µ–Ω–∏–µ –≤ –≥–æ–Ω–∫–µ (1 –∑–∞ –ø–µ—Ä–≤–æ–µ –º–µ—Å—Ç–æ, 3 –∑–∞ —Ç—Ä–µ—Ç—å–µ –∏ —Ç.–¥.).

–û–Ω –Ω–∞–ø–∏—Å–∞–ª —Ñ—É–Ω–∫—Ü–∏—é, –æ–ø–∏—Å–∞–Ω–Ω—É—é –Ω–∏–∂–µ, —á—Ç–æ–±—ã –æ–Ω–∞ –±—Ä–∞–ª–∞ —Ç–∞–∫–æ–π —Å–ø–∏—Å–æ–∫ –∏ –≤–æ–∑–≤—Ä–∞—â–∞–ª–∞ —Å–ª–æ–≤–∞—Ä—å, –æ—Ç–æ–±—Ä–∞–∂–∞—é—â–∏–π –∫–∞–∂–¥—ã–π —ç–ª–µ–º–µ–Ω—Ç –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å —Ç–µ–º, —Å–∫–æ–ª—å–∫–æ —Ä–∞–∑ –æ–Ω –±—ã–ª –≤—ã–±—Ä–∞–Ω —É—á–∞—Å—Ç–Ω–∏–∫–∞–º–∏, –∑–∞–Ω—è–≤—à–∏–º–∏ –ø–µ—Ä–≤–æ–µ –º–µ—Å—Ç–æ.

```python
def best_items(racers):
    """–ü–æ–ª—É—á–∏–≤ —Å–ø–∏—Å–æ–∫ —Å–ª–æ–≤–∞—Ä–µ–π –¥–ª—è –≥–æ–Ω—â–∏–∫–æ–≤, –≤–µ—Ä–Ω–∏—Ç–µ –≤ —Å–ª–æ–≤–∞—Ä—å —ç–ª–µ–º–µ–Ω—Ç—ã, —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É
    —Å–ª—É—á–∞–µ–≤, –∫–æ–≥–¥–∞ —ç—Ç–∏ —ç–ª–µ–º–µ–Ω—Ç—ã –±—ã–ª–∏ —Å–æ–±—Ä–∞–Ω—ã –≥–æ–Ω—â–∏–∫–∞–º–∏, –∑–∞–Ω—è–≤—à–∏–º–∏ –ø–µ—Ä–≤–æ–µ –º–µ—Å—Ç–æ.
    """
    winner_item_counts = {}
    for i in range(len(racers)):
        # –ü–µ—Ä–≤—ã–π —Å–ª–æ–≤–∞—Ä—å –≥–æ–Ω—â–∏–∫–∞
        racer = racers[i]
        # –ù–∞—Å –∏–Ω—Ç–µ—Ä–µ—Å—É—é—Ç —Ç–æ–ª—å–∫–æ —Ç–µ –≥–æ–Ω—â–∏–∫–∏, –∫–æ—Ç–æ—Ä—ã–µ —Ñ–∏–Ω–∏—à–∏—Ä–æ–≤–∞–ª–∏ –ø–µ—Ä–≤—ã–º–∏.
        if racer['finish'] == 1:
            for i in racer['items']:
                # –î–æ–±–∞–≤—å—Ç–µ –µ–¥–∏–Ω–∏—Ü—É –∫ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –¥–ª—è —ç—Ç–æ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞ (–ø—Ä–∏ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ—Å—Ç–∏ –¥–æ–±–∞–≤—å—Ç–µ –µ–≥–æ –≤ dict)
                if i not in winner_item_counts:
                    winner_item_counts[i] = 0
                winner_item_counts[i] += 1

        # –ü—Ä–æ–±–ª–µ–º—ã —Å –∫–∞—á–µ—Å—Ç–≤–æ–º –¥–∞–Ω–Ω—ã—Ö :/ –í—ã–≤–æ–¥ –ø—Ä–µ–¥—É–ø—Ä–µ–∂–¥–µ–Ω–∏—è –æ –≥–æ–Ω—â–∏–∫–∞—Ö –±–µ–∑ —É–∫–∞–∑–∞–Ω–∏—è –∏–º–µ–Ω–∏. –ú—ã –ø–æ–∑–∞–±–æ—Ç–∏–º—Å—è –æ–± —ç—Ç–æ–º –ø–æ–∑–∂–µ.
        if racer['name'] is None:
            print("WARNING: Encountered racer with unknown name on iteration {}/{} (racer = {})".format(
                i+1, len(racers), racer['name'])
                 )
    return winner_item_counts
```

–û–Ω –ø–æ–ø—Ä–æ–±–æ–≤–∞–ª —ç—Ç–æ –Ω–∞ –ø—Ä–∏–º–µ—Ä–µ –Ω–µ–±–æ–ª—å—à–æ–≥–æ —Å–ø–∏—Å–∫–∞ –ø—Ä–∏–º–µ—Ä–æ–≤, –ø—Ä–∏–≤–µ–¥–µ–Ω–Ω–æ–≥–æ –≤—ã—à–µ, –∏, –ø–æ—Ö–æ–∂–µ, –≤—Å–µ —Å—Ä–∞–±–æ—Ç–∞–ª–æ –ø—Ä–∞–≤–∏–ª—å–Ω–æ:

```python
sample = [
    {'name': 'Peach', 'items': ['green shell', 'banana', 'green shell',], 'finish': 3},
    {'name': 'Bowser', 'items': ['green shell',], 'finish': 1},
    {'name': None, 'items': ['mushroom',], 'finish': 2},
    {'name': 'Toad', 'items': ['green shell', 'mushroom'], 'finish': 1},
]
best_items(sample)
```

–ú–æ–∂–µ—Ç–µ –ª–∏ –≤—ã –¥–æ–≥–∞–¥–∞—Ç—å—Å—è, –ø–æ—á–µ–º—É? –ü–æ–ø—Ä–æ–±—É–π—Ç–µ –∑–∞–ø—É—Å—Ç–∏—Ç—å –ø—Ä–∏–≤–µ–¥–µ–Ω–Ω—É—é –Ω–∏–∂–µ —è—á–µ–π–∫—É –∫–æ–¥–∞, —á—Ç–æ–±—ã —É–≤–∏–¥–µ—Ç—å —Å–æ–æ–±—â–µ–Ω–∏–µ –æ–± –æ—à–∏–±–∫–µ, –∫–æ—Ç–æ—Ä–æ–µ –ø–æ–ª—É—á–∞–µ—Ç –õ—É–∏–¥–∂–∏. –ö–∞–∫ —Ç–æ–ª—å–∫–æ –≤—ã –≤—ã—è–≤–∏—Ç–µ –æ—à–∏–±–∫—É, –∏—Å–ø—Ä–∞–≤—å—Ç–µ –µ–µ –≤ —è—á–µ–π–∫–µ –Ω–∏–∂–µ (—á—Ç–æ–±—ã –ø—Ä–æ–≥—Ä–∞–º–º–∞ —Ä–∞–±–æ—Ç–∞–ª–∞ –±–µ–∑ –æ—à–∏–±–æ–∫).

–ü–æ–¥—Å–∫–∞–∑–∫–∞: –æ—à–∏–±–∫–∞ –õ—É–∏–¥–∂–∏ –ø–æ—Ö–æ–∂–∞ –Ω–∞ —Ç—É, —Å –∫–æ—Ç–æ—Ä–æ–π –º—ã —Å—Ç–æ–ª–∫–Ω—É–ª–∏—Å—å –≤ —Ä—É–∫–æ–≤–æ–¥—Å—Ç–≤–µ, –∫–æ–≥–¥–∞ –≥–æ–≤–æ—Ä–∏–ª–∏ –æ–± –∏–º–ø–æ—Ä—Ç–µ star.

```python
# –ò–º–ø–æ—Ä—Ç–∏—Ä—É–π—Ç–µ –ø–æ–ª–Ω—ã–π –Ω–∞–±–æ—Ä –≥–æ–Ω–æ—á–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö –õ—É–∏–¥–∂–∏.
from learntools.python.luigi_analysis import full_dataset

# –ò—Å–ø—Ä–∞–≤—å –º–µ–Ω—è!
def best_items(racers):
    winner_item_counts = {}
    for i in range(len(racers)):
        # –ü–µ—Ä–≤—ã–π —Å–ª–æ–≤–∞—Ä—å –≥–æ–Ω—â–∏–∫–∞
        racer = racers[i]
        # –ù–∞—Å –∏–Ω—Ç–µ—Ä–µ—Å—É—é—Ç —Ç–æ–ª—å–∫–æ —Ç–µ –≥–æ–Ω—â–∏–∫–∏, –∫–æ—Ç–æ—Ä—ã–µ —Ñ–∏–Ω–∏—à–∏—Ä–æ–≤–∞–ª–∏ –ø–µ—Ä–≤—ã–º–∏.
        if racer['finish'] == 1:
            for i in racer['items']:
                # –î–æ–±–∞–≤—å—Ç–µ –µ–¥–∏–Ω–∏—Ü—É –∫ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –¥–ª—è —ç—Ç–æ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞ (–ø—Ä–∏ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ—Å—Ç–∏ –¥–æ–±–∞–≤—å—Ç–µ –µ–≥–æ –≤ dict)
                if i not in winner_item_counts:
                    winner_item_counts[i] = 0
                winner_item_counts[i] += 1

        # –ü—Ä–æ–±–ª–µ–º—ã —Å –∫–∞—á–µ—Å—Ç–≤–æ–º –¥–∞–Ω–Ω—ã—Ö :/ –í—ã–≤–æ–¥ –ø—Ä–µ–¥—É–ø—Ä–µ–∂–¥–µ–Ω–∏—è –æ –≥–æ–Ω—â–∏–∫–∞—Ö –±–µ–∑ —É–∫–∞–∑–∞–Ω–∏—è –∏–º–µ–Ω–∏. –ú—ã –ø–æ–∑–∞–±–æ—Ç–∏–º—Å—è –æ–± —ç—Ç–æ–º –ø–æ–∑–∂–µ.
        if racer['name'] is None:
            print("WARNING: Encountered racer with unknown name on iteration {}/{} (racer = {})".format(
                i+1, len(racers), racer['name'])
                 )
    return winner_item_counts

# –ü–æ–ø—Ä–æ–±—É–π—Ç–µ –ø—Ä–æ–∞–Ω–∞–ª–∏–∑–∏—Ä–æ–≤–∞—Ç—å –∏–º–ø–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –ø–æ–ª–Ω—ã–π –Ω–∞–±–æ—Ä –¥–∞–Ω–Ω—ã—Ö
best_items(full_dataset)
```
