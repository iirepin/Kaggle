```python
def fashionably_late(arrivals, name):
    """–ü–æ–ª—É—á–∏–≤ —Å–ø–∏—Å–æ–∫ –ø—Ä–∏–≥–ª–∞—à–µ–Ω–Ω—ã—Ö –Ω–∞ –≤–µ—á–µ—Ä–∏–Ω–∫—É –∏ —É–∫–∞–∑–∞–≤ –∏–º—è, —É—Ç–æ—á–Ω–∏—Ç–µ, –æ–ø–æ–∑–¥–∞–ª –ª–∏ –≥–æ—Å—Ç—å —Å —Ç–∞–∫–∏–º
–∏–º–µ–Ω–µ–º –ø–æ –ø—Ä–∞–≤–∏–ª–∞–º –ø—Ä–∏–ª–∏—á–∏—è.
    """
    list_fashionably = arrivals[round(len(arrivals) / 2) : -1]
    if name in list_fashionably:
        return True
    else:
        return False
fashionably_late(arrivals, "–û—É—ç–Ω")
```
