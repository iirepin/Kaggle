# –°—Ç—Ä–æ–∫–∏ –∏ —Å–ª–æ–≤–∞—Ä–∏
–†–∞–±–æ—Ç–∞ —Å–æ —Å—Ç—Ä–æ–∫–∞–º–∏ –∏ —Å–ª–æ–≤–∞—Ä—è–º–∏, –¥–≤—É–º—è –æ—Å–Ω–æ–≤–Ω—ã–º–∏ —Ç–∏–ø–∞–º–∏ –¥–∞–Ω–Ω—ã—Ö Python

# –°—Ç—Ä—É–Ω–Ω—ã–µ¬∂
–ï–¥–∏–Ω—Å—Ç–≤–µ–Ω–Ω–æ–µ, –≤ —á–µ–º —è–∑—ã–∫ Python –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ –ø—Ä–µ—É—Å–ø–µ–≤–∞–µ—Ç, - —ç—Ç–æ –≤ –º–∞–Ω–∏–ø—É–ª–∏—Ä–æ–≤–∞–Ω–∏–∏ —Å—Ç—Ä–æ–∫–∞–º–∏. –í —ç—Ç–æ–º —Ä–∞–∑–¥–µ–ª–µ –±—É–¥—É—Ç —Ä–∞—Å—Å–º–æ—Ç—Ä–µ–Ω—ã –Ω–µ–∫–æ—Ç–æ—Ä—ã–µ –≤—Å—Ç—Ä–æ–µ–Ω–Ω—ã–µ –≤ Python —Å—Ç—Ä–æ–∫–æ–≤—ã–µ –º–µ—Ç–æ–¥—ã –∏ –æ–ø–µ—Ä–∞—Ü–∏–∏ —Ñ–æ—Ä–º–∞—Ç–∏—Ä–æ–≤–∞–Ω–∏—è.

–¢–∞–∫–∏–µ —à–∞–±–ª–æ–Ω—ã –º–∞–Ω–∏–ø—É–ª–∏—Ä–æ–≤–∞–Ω–∏—è —Å—Ç—Ä–æ–∫–∞–º–∏ —á–∞—Å—Ç–æ –≤—Å—Ç—Ä–µ—á–∞—é—Ç—Å—è –≤ –∫–æ–Ω—Ç–µ–∫—Å—Ç–µ —Ä–∞–±–æ—Ç—ã —Å –¥–∞–Ω–Ω—ã–º–∏.

# –°–∏–Ω—Ç–∞–∫—Å–∏—Å —Å—Ç—Ä–æ–∫
–í—ã —É–∂–µ –≤–∏–¥–µ–ª–∏ –º–Ω–æ–∂–µ—Å—Ç–≤–æ —Å—Ç—Ä–æ–∫ –≤ –ø—Ä–∏–º–µ—Ä–∞—Ö –Ω–∞ –ø—Ä–µ–¥—ã–¥—É—â–∏—Ö —É—Ä–æ–∫–∞—Ö, –Ω–æ –ø—Ä–æ—Å—Ç–æ –Ω–∞–ø–æ–º–Ω—é, —á—Ç–æ —Å—Ç—Ä–æ–∫–∏ –≤ Python –º–æ–≥—É—Ç –±—ã—Ç—å –æ–ø—Ä–µ–¥–µ–ª–µ–Ω—ã —Å –ø–æ–º–æ—â—å—é –æ–¥–∏–Ω–∞—Ä–Ω—ã—Ö –∏–ª–∏ –¥–≤–æ–π–Ω—ã—Ö –∫–∞–≤—ã—á–µ–∫. –§—É–Ω–∫—Ü–∏–æ–Ω–∞–ª—å–Ω–æ –æ–Ω–∏ —ç–∫–≤–∏–≤–∞–ª–µ–Ω—Ç–Ω—ã.

```python
x = 'Pluto is a planet'
y = "Pluto is a planet"
x == y
```

–î–≤–æ–π–Ω—ã–µ –∫–∞–≤—ã—á–∫–∏ —É–¥–æ–±–Ω—ã, –µ—Å–ª–∏ –≤–∞—à–∞ —Å—Ç—Ä–æ–∫–∞ —Å–æ–¥–µ—Ä–∂–∏—Ç —Å–∏–º–≤–æ–ª –æ–¥–∏–Ω–∞—Ä–Ω–æ–π –∫–∞–≤—ã—á–∫–∏ (–Ω–∞–ø—Ä–∏–º–µ—Ä, –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è—é—â–∏–π –∞–ø–æ—Å—Ç—Ä–æ—Ñ).

–ê–Ω–∞–ª–æ–≥–∏—á–Ω–æ, –ª–µ–≥–∫–æ —Å–æ–∑–¥–∞—Ç—å —Å—Ç—Ä–æ–∫—É, —Å–æ–¥–µ—Ä–∂–∞—â—É—é –¥–≤–æ–π–Ω—ã–µ –∫–∞–≤—ã—á–∫–∏, –µ—Å–ª–∏ –∑–∞–∫–ª—é—á–∏—Ç—å –µ–µ –≤ –æ–¥–∏–Ω–∞—Ä–Ω—ã–µ –∫–∞–≤—ã—á–∫–∏:

```python
print("Pluto's a planet!")
print('My dog is named "Pluto"')
```

–ï—Å–ª–∏ –º—ã –ø–æ–ø—ã—Ç–∞–µ–º—Å—è –ø–æ–º–µ—Å—Ç–∏—Ç—å —Å–∏–º–≤–æ–ª –æ–¥–∏–Ω–∞—Ä–Ω–æ–π –∫–∞–≤—ã—á–∫–∏ –≤ —Å—Ç—Ä–æ–∫—É, –∑–∞–∫–ª—é—á–µ–Ω–Ω—É—é –≤ –æ–¥–∏–Ω–∞—Ä–Ω—ã–µ –∫–∞–≤—ã—á–∫–∏, Python –∑–∞–ø—É—Ç–∞–µ—Ç—Å—è:

```python
'Pluto's a planet!'
```

–ú—ã –º–æ–∂–µ–º –∏—Å–ø—Ä–∞–≤–∏—Ç—å —ç—Ç–æ, "—ç–∫—Ä–∞–Ω–∏—Ä—É—è" –æ–¥–∏–Ω–∞—Ä–Ω—É—é –∫–∞–≤—ã—á–∫—É –æ–±—Ä–∞—Ç–Ω–æ–π –∫–æ—Å–æ–π —á–µ—Ä—Ç–æ–π.

```python
'Pluto\'s a planet!'
```

–í —Ç–∞–±–ª–∏—Ü–µ –Ω–∏–∂–µ –ø—Ä–∏–≤–µ–¥–µ–Ω—ã –Ω–µ–∫–æ—Ç–æ—Ä—ã–µ –≤–∞–∂–Ω—ã–µ –≤–∞—Ä–∏–∞–Ω—Ç—ã –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è —Å–∏–º–≤–æ–ª–∞ –æ–±—Ä–∞—Ç–Ω–æ–π –∫–æ—Å–æ–π —á–µ—Ä—Ç—ã.


–ü–æ—Å–ª–µ–¥–Ω—è—è –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å, \n, –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è–µ—Ç —Å–æ–±–æ–π —Å–∏–º–≤–æ–ª –Ω–æ–≤–æ–π —Å—Ç—Ä–æ–∫–∏. –≠—Ç–æ –∑–∞—Å—Ç–∞–≤–ª—è–µ—Ç Python –Ω–∞—á–∏–Ω–∞—Ç—å –Ω–æ–≤—É—é —Å—Ç—Ä–æ–∫—É.

```python
hello = "hello\nworld"
print(hello)
```

–ö—Ä–æ–º–µ —Ç–æ–≥–æ, —Å–∏–Ω—Ç–∞–∫—Å–∏—Å —Ç—Ä–æ–π–Ω—ã—Ö –∫–∞–≤—ã—á–µ–∫ Python –¥–ª—è —Å—Ç—Ä–æ–∫ –ø–æ–∑–≤–æ–ª—è–µ—Ç –Ω–∞–º –≤–∫–ª—é—á–∞—Ç—å –Ω–æ–≤—ã–µ —Å—Ç—Ä–æ–∫–∏ –±—É–∫–≤–∞–ª—å–Ω–æ (—Ç.–µ. –ø—Ä–æ—Å—Ç–æ –Ω–∞–∂–∏–º–∞—è "Enter" –Ω–∞ –∫–ª–∞–≤–∏–∞—Ç—É—Ä–µ, –∞ –Ω–µ –∏—Å–ø–æ–ª—å–∑—É—è —Å–ø–µ—Ü–∏–∞–ª—å–Ω—É—é –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å "\n"). –ú—ã —É–∂–µ –≤–∏–¥–µ–ª–∏ —ç—Ç–æ –≤ —Å—Ç—Ä–æ–∫–∞—Ö –¥–æ–∫—É–º–µ–Ω—Ç–∞—Ü–∏–∏, –∫–æ—Ç–æ—Ä—ã–µ –º—ã –∏—Å–ø–æ–ª—å–∑—É–µ–º –¥–ª—è –¥–æ–∫—É–º–µ–Ω—Ç–∏—Ä–æ–≤–∞–Ω–∏—è –Ω–∞—à–∏—Ö —Ñ—É–Ω–∫—Ü–∏–π, –Ω–æ –º—ã –º–æ–∂–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –∏—Ö –≤–µ–∑–¥–µ, –≥–¥–µ –∑–∞—Ö–æ—Ç–∏–º –æ–ø—Ä–µ–¥–µ–ª–∏—Ç—å —Å—Ç—Ä–æ–∫—É.

```python
triplequoted_hello = """hello
world"""
print(triplequoted_hello)
triplequoted_hello == hello
```

–§—É–Ω–∫—Ü–∏—è print() –∞–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∏ –¥–æ–±–∞–≤–ª—è–µ—Ç —Å–∏–º–≤–æ–ª –Ω–æ–≤–æ–π —Å—Ç—Ä–æ–∫–∏, –µ—Å–ª–∏ –º—ã –Ω–µ —É–∫–∞–∂–µ–º –∑–Ω–∞—á–µ–Ω–∏–µ –¥–ª—è –∫–ª—é—á–µ–≤–æ–≥–æ —Å–ª–æ–≤–∞ –∞—Ä–≥—É–º–µ–Ω—Ç end, –æ—Ç–ª–∏—á–Ω–æ–µ –æ—Ç –∑–Ω–∞—á–µ–Ω–∏—è –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é '\n':

```python
print("hello")
print("world")
print("hello", end='')
print("pluto", end='')
```

# –°—Ç—Ä–æ–∫–∏ - —ç—Ç–æ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏¬∂
–°—Ç—Ä–æ–∫–∏ –º–æ–∂–Ω–æ —Ä–∞—Å—Å–º–∞—Ç—Ä–∏–≤–∞—Ç—å –∫–∞–∫ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ —Å–∏–º–≤–æ–ª–æ–≤. –ü–æ—á—Ç–∏ –≤—Å–µ, —á—Ç–æ, –∫–∞–∫ –º—ã –≤–∏–¥–µ–ª–∏, –º—ã –º–æ–∂–µ–º —Å–¥–µ–ª–∞—Ç—å —Å–æ —Å–ø–∏—Å–∫–æ–º, –º—ã —Ç–∞–∫–∂–µ –º–æ–∂–µ–º —Å–¥–µ–ª–∞—Ç—å —Å–æ —Å—Ç—Ä–æ–∫–æ–π.

```python
# –ò–Ω–¥–µ–∫—Å–∞—Ü–∏—è
planet = 'Pluto'
planet[0]
```

```python
# –ù–∞—Ä–µ–∑–∫–∞
planet[-3:]
```

```python
# –ö–∞–∫–æ–π –¥–ª–∏–Ω—ã —ç—Ç–∞ —Å—Ç—Ä–æ–∫–∞?
len(planet)
```

```python
# –î–∞, –º—ã –¥–∞–∂–µ –º–æ–∂–µ–º –æ–±–æ–π—Ç–∏ –∏—Ö –ø–æ –∫—Ä—É–≥—É.
[char+'! ' for char in planet]
```

–ù–æ –≥–ª–∞–≤–Ω–æ–µ, —á–µ–º –æ–Ω–∏ –æ—Ç–ª–∏—á–∞—é—Ç—Å—è –æ—Ç —Å–ø–∏—Å–∫–æ–≤, - —ç—Ç–æ —Ç–æ, —á—Ç–æ –æ–Ω–∏ –Ω–µ–∏–∑–º–µ–Ω—è–µ–º—ã. –ú—ã –Ω–µ –º–æ–∂–µ–º –∏—Ö –∏–∑–º–µ–Ω—è—Ç—å.

```python
planet[0] = 'B'
# planet.append —Ç–æ–∂–µ –Ω–µ —Ä–∞–±–æ—Ç–∞–µ—Ç
```

# –°—Ç—Ä–æ–∫–æ–≤—ã–µ –º–µ—Ç–æ–¥—ã
–ö–∞–∫ –∏ list, —Ç–∏–ø str –∏–º–µ–µ—Ç –º–Ω–æ–∂–µ—Å—Ç–≤–æ –æ—á–µ–Ω—å –ø–æ–ª–µ–∑–Ω—ã—Ö –º–µ—Ç–æ–¥–æ–≤. –ó–¥–µ—Å—å —è –ø–æ–∫–∞–∂—É –ª–∏—à—å –Ω–µ—Å–∫–æ–ª—å–∫–æ –ø—Ä–∏–º–µ—Ä–æ–≤.

```python
# –í–°–ï –ó–ê–ì–õ–ê–í–ù–´–ï –ë–£–ö–í–´
claim = "Pluto is a planet!"
claim.upper()
```

```python
# –≤—Å–µ –≤ –Ω–∏–∂–Ω–µ–º —Ä–µ–≥–∏—Å—Ç—Ä–µ
claim.lower()
```

```python
# –ü–æ–∏—Å–∫ –ø–µ—Ä–≤–æ–≥–æ –∏–Ω–¥–µ–∫—Å–∞ –ø–æ–¥—Å—Ç—Ä–æ–∫–∏
claim.index('plan')
```

```python
claim.startswith(planet)
```

```python
# –Ω–µ–≤–µ—Ä–Ω–æ –∏–∑-–∑–∞ –æ—Ç—Å—É—Ç—Å—Ç–≤–∏—è –≤–æ—Å–∫–ª–∏—Ü–∞—Ç–µ–ª—å–Ω–æ–≥–æ –∑–Ω–∞–∫–∞
claim.endswith('planet')
```

–ü–µ—Ä–µ–∫–ª—é—á–µ–Ω–∏–µ –º–µ–∂–¥—É —Å—Ç—Ä–æ–∫–∞–º–∏ –∏ —Å–ø–∏—Å–∫–∞–º–∏: .split() –∏ .join()¬∂

—Ñ—É–Ω–∫—Ü–∏—è str.split() –ø—Ä–µ–æ–±—Ä–∞–∑—É–µ—Ç —Å—Ç—Ä–æ–∫—É –≤ —Å–ø–∏—Å–æ–∫ —Å—Ç—Ä–æ–∫ –º–µ–Ω—å—à–µ–≥–æ —Ä–∞–∑–º–µ—Ä–∞, –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é —Ä–∞–∑–±–∏–≤–∞—è –µ–µ –Ω–∞ –ø—Ä–æ–±–µ–ª—ã. –≠—Ç–æ –æ—á–µ–Ω—å —É–¥–æ–±–Ω–æ –¥–ª—è –ø–µ—Ä–µ—Ö–æ–¥–∞ –æ—Ç –æ–¥–Ω–æ–π –±–æ–ª—å—à–æ–π —Å—Ç—Ä–æ–∫–∏ –∫ —Å–ø–∏—Å–∫—É —Å–ª–æ–≤.

```python
words = claim.split()
words
```

–ò–Ω–æ–≥–¥–∞ –≤–∞–º –∑–∞—Ö–æ—á–µ—Ç—Å—è –≤—ã–¥–µ–ª–∏—Ç—å —á—Ç–æ-—Ç–æ –¥—Ä—É–≥–æ–µ, –∫—Ä–æ–º–µ –ø—Ä–æ–±–µ–ª–æ–≤:

```python
datestr = '1956-01-31'
year, month, day = datestr.split('-')
```

str.join() –≤–µ–¥–µ—Ç –Ω–∞—Å –≤ –¥—Ä—É–≥–æ–º –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏–∏, –æ–±—ä–µ–¥–∏–Ω—è—è —Å–ø–∏—Å–æ–∫ —Å—Ç—Ä–æ–∫ –≤ –æ–¥–Ω—É –¥–ª–∏–Ω–Ω—É—é —Å—Ç—Ä–æ–∫—É, –∏—Å–ø–æ–ª—å–∑—É—è —Å—Ç—Ä–æ–∫—É, –¥–ª—è –∫–æ—Ç–æ—Ä–æ–π –æ–Ω–∞ –±—ã–ª–∞ –≤—ã–∑–≤–∞–Ω–∞, –≤ –∫–∞—á–µ—Å—Ç–≤–µ —Ä–∞–∑–¥–µ–ª–∏—Ç–µ–ª—è.

```python
'/'.join([month, day, year])
```

```python
# –î–∞, –º—ã –º–æ–∂–µ–º –ø–æ–º–µ—Å—Ç–∏—Ç—å —Å–∏–º–≤–æ–ª—ã —é–Ω–∏–∫–æ–¥–∞ –ø—Ä—è–º–æ –≤ –Ω–∞—à–∏ —Å—Ç—Ä–æ–∫–æ–≤—ã–µ –ª–∏—Ç–µ—Ä–∞–ª—ã :)
' üëè '.join([word.upper() for word in words])
```

–ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ —Å—Ç—Ä–æ–∫ —Å –ø–æ–º–æ—â—å—é .format()¬∂

Python –ø–æ–∑–≤–æ–ª—è–µ—Ç –Ω–∞–º –æ–±—ä–µ–¥–∏–Ω—è—Ç—å —Å—Ç—Ä–æ–∫–∏ —Å –ø–æ–º–æ—â—å—é –æ–ø–µ—Ä–∞—Ç–æ—Ä–∞ +.

```python
planet + ', we miss you.'
```

–ï—Å–ª–∏ –º—ã —Ö–æ—Ç–∏–º –¥–æ–±–∞–≤–∏—Ç—å –∫–∞–∫–∏–µ-–ª–∏–±–æ –Ω–µ—Å—Ç—Ä–æ–∫–æ–≤—ã–µ –æ–±—ä–µ–∫—Ç—ã, –º—ã –¥–æ–ª–∂–Ω—ã –±—ã—Ç—å –æ—Å—Ç–æ—Ä–æ–∂–Ω—ã –ø—Ä–∏ –≤—ã–∑–æ–≤–µ str() –¥–ª—è –Ω–∏—Ö –≤ –ø–µ—Ä–≤—É—é –æ—á–µ—Ä–µ–¥—å

```python
position = 9
planet + ", you'll always be the " + position + "th planet to me."
```

```python
planet + ", you'll always be the " + str(position) + "th planet to me."
```

–≠—Ç–æ —Å—Ç–∞–Ω–æ–≤–∏—Ç—Å—è —Ç—Ä—É–¥–Ω–æ —á–∏—Ç–∞—Ç—å –∏ —Ä–∞–∑–¥—Ä–∞–∂–∞–µ—Ç, –∫–æ–≥–¥–∞ –ø—Ä–∏—Ö–æ–¥–∏—Ç—Å—è –≤–≤–æ–¥–∏—Ç—å. str.format() –Ω–∞ –ø–æ–º–æ—â—å.

```python
"{}, you'll always be the {}th planet to me.".format(planet, position)
```

–ù–∞–º–Ω–æ–≥–æ –ø—Ä–æ—â–µ! –ú—ã –≤—ã–∑—ã–≤–∞–µ–º .format() –¥–ª—è "—Å—Ç—Ä–æ–∫–∏ —Ñ–æ—Ä–º–∞—Ç–∞", –≥–¥–µ –∑–Ω–∞—á–µ–Ω–∏—è Python, –∫–æ—Ç–æ—Ä—ã–µ –º—ã —Ö–æ—Ç–∏–º –≤—Å—Ç–∞–≤–∏—Ç—å, –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω—ã —Å –ø–æ–º–æ—â—å—é {}-–∑–∞–ø–æ–ª–Ω–∏—Ç–µ–ª–µ–π.

–û–±—Ä–∞—Ç–∏—Ç–µ –≤–Ω–∏–º–∞–Ω–∏–µ, —á—Ç–æ –Ω–∞–º –¥–∞–∂–µ –Ω–µ –ø—Ä–∏—à–ª–æ—Å—å –≤—ã–∑—ã–≤–∞—Ç—å str() –¥–ª—è –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è –ø–æ–∑–∏—Ü–∏–∏ –∏–∑ int. —Ñ—É–Ω–∫—Ü–∏—è format() –ø–æ–∑–∞–±–æ—Ç–∏—Ç—Å—è –æ–± —ç—Ç–æ–º –∑–∞ –Ω–∞—Å.

–ï—Å–ª–∏ –±—ã —ç—Ç–æ –±—ã–ª–æ –≤—Å–µ, —á—Ç–æ –¥–µ–ª–∞–ª–∞ —Ñ—É–Ω–∫—Ü–∏—è format(), –æ–Ω–∞ –≤—Å–µ —Ä–∞–≤–Ω–æ –±—ã–ª–∞ –±—ã –Ω–µ–≤–µ—Ä–æ—è—Ç–Ω–æ –ø–æ–ª–µ–∑–Ω–∞. –ù–æ, –∫–∞–∫ –æ–∫–∞–∑–∞–ª–æ—Å—å, –æ–Ω–∞ –º–æ–∂–µ—Ç —Å–¥–µ–ª–∞—Ç—å –≥–æ—Ä–∞–∑–¥–æ –±–æ–ª—å—à–µ. –í–æ—Ç –ª–∏—à—å –∫—Ä–∞—Ç–∫–∏–π –æ–±–∑–æ—Ä:

```python
pluto_mass = 1.303 * 10**22
earth_mass = 5.9722 * 10**24
population = 52910390
#         2 decimal points   3 decimal points, format as percent     separate with commas
"{} weighs about {:.2} kilograms ({:.3%} of Earth's mass). It is home to {:,} Plutonians.".format(
    planet, pluto_mass, pluto_mass / earth_mass, population,
)
```

```python
# –°—Å—ã–ª–∫–∞ –Ω–∞ –∞—Ä–≥—É–º–µ–Ω—Ç—ã format() –ø–æ –∏–Ω–¥–µ–∫—Å—É, –Ω–∞—á–∏–Ω–∞—è —Å 0
s = """Pluto's a {0}.
No, it's a {1}.
{0}!
{1}!""".format('planet', 'dwarf planet')
print(s)
```

–í–µ—Ä–æ—è—Ç–Ω–æ, –≤—ã –º–æ–≥–ª–∏ –±—ã –Ω–∞–ø–∏—Å–∞—Ç—å –Ω–µ–±–æ–ª—å—à—É—é –∫–Ω–∏–≥—É —Ç–æ–ª—å–∫–æ –≤ —Ñ–æ—Ä–º–∞—Ç–µ str., –ø–æ—ç—Ç–æ–º—É —è –æ—Å—Ç–∞–Ω–æ–≤–ª—é—Å—å –∑–¥–µ—Å—å –∏ —É–∫–∞–∂—É –≤–∞–º –Ω–∞ pyformat.info –∏ –æ—Ñ–∏—Ü–∏–∞–ª—å–Ω—ã–µ –¥–æ–∫—É–º–µ–Ω—Ç—ã –¥–ª—è –¥–∞–ª—å–Ω–µ–π—à–µ–≥–æ —á—Ç–µ–Ω–∏—è.

# –°–ª–æ–≤–∞—Ä–∏¬∂
–°–ª–æ–≤–∞—Ä–∏ - —ç—Ç–æ –≤—Å—Ç—Ä–æ–µ–Ω–Ω–∞—è —Å—Ç—Ä—É–∫—Ç—É—Ä–∞ –¥–∞–Ω–Ω—ã—Ö Python –¥–ª—è —Å–æ–ø–æ—Å—Ç–∞–≤–ª–µ–Ω–∏—è –∫–ª—é—á–µ–π —Å–æ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏.

```python
numbers = {'one':1, 'two':2, 'three':3}
```

–í –¥–∞–Ω–Ω–æ–º —Å–ª—É—á–∞–µ "–æ–¥–∏–Ω", "–¥–≤–∞" –∏ "—Ç—Ä–∏" —è–≤–ª—è—é—Ç—Å—è –∫–ª—é—á–∞–º–∏, –∞ 1, 2 –∏ 3 - –∏—Ö —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏.

–î–æ—Å—Ç—É–ø –∫ –∑–Ω–∞—á–µ–Ω–∏—è–º –æ—Å—É—â–µ—Å—Ç–≤–ª—è–µ—Ç—Å—è —Å –ø–æ–º–æ—â—å—é —Å–∏–Ω—Ç–∞–∫—Å–∏—Å–∞ –∫–≤–∞–¥—Ä–∞—Ç–Ω—ã—Ö —Å–∫–æ–±–æ–∫, –∞–Ω–∞–ª–æ–≥–∏—á–Ω–æ–≥–æ –∏–Ω–¥–µ–∫—Å–∞—Ü–∏–∏ –≤ —Å–ø–∏—Å–∫–∞—Ö –∏ —Å—Ç—Ä–æ–∫–∞—Ö.

```python
numbers['one']
```

–ú—ã –º–æ–∂–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —Ç–æ—Ç –∂–µ —Å–∏–Ω—Ç–∞–∫—Å–∏—Å –¥–ª—è –¥–æ–±–∞–≤–ª–µ–Ω–∏—è –¥—Ä—É–≥–æ–π –ø–∞—Ä—ã –∫–ª—é—á-–∑–Ω–∞—á–µ–Ω–∏–µ

```python
numbers['eleven'] = 11
numbers
```

–ò–ª–∏ –∏–∑–º–µ–Ω–∏—Ç—å –∑–Ω–∞—á–µ–Ω–∏–µ, —Å–≤—è–∑–∞–Ω–Ω–æ–µ —Å —Å—É—â–µ—Å—Ç–≤—É—é—â–∏–º –∫–ª—é—á–æ–º

```python
numbers['one'] = 'Pluto'
numbers
```

–í Python –µ—Å—Ç—å –ø–æ–Ω–∏–º–∞–Ω–∏–µ –ø–æ —Å–ª–æ–≤–∞—Ä—é —Å —Å–∏–Ω—Ç–∞–∫—Å–∏—Å–æ–º, –ø–æ—Ö–æ–∂–∏–º –Ω–∞ –ø–æ–Ω–∏–º–∞–Ω–∏–µ –ø–æ —Å–ø–∏—Å–∫—É, –∫–æ—Ç–æ—Ä–æ–µ –º—ã –≤–∏–¥–µ–ª–∏ –≤ –ø—Ä–µ–¥—ã–¥—É—â–µ–º —Ä—É–∫–æ–≤–æ–¥—Å—Ç–≤–µ.

```python
planets = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Uranus', 'Neptune']
planet_to_initial = {planet: planet[0] for planet in planets}
planet_to_initial
```

The in operator tells us whether something is a key in the dictionary

```python
'Saturn' in planet_to_initial
```

```python
'Betelgeuse' in planet_to_initial
```

–¶–∏–∫–ª for –ø–æ —Å–ª–æ–≤–∞—Ä—é –±—É–¥–µ—Ç –≤—ã–ø–æ–ª–Ω—è—Ç—å —Ü–∏–∫–ª –ø–æ –µ–≥–æ –∫–ª—é—á–∞–º

```python
for k in numbers:
    print("{} = {}".format(k, numbers[k]))
```

–ú—ã –º–æ–∂–µ–º –ø–æ–ª—É—á–∏—Ç—å –¥–æ—Å—Ç—É–ø –∫ –∫–æ–ª–ª–µ–∫—Ü–∏–∏ –≤—Å–µ—Ö –∫–ª—é—á–µ–π –∏–ª–∏ –≤—Å–µ—Ö –∑–Ω–∞—á–µ–Ω–∏–π —Å –ø–æ–º–æ—â—å—é dict.keys() –∏ dict.values() —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ.

```python
# –°–æ–±–µ—Ä–∏—Ç–µ –≤—Å–µ –∏–Ω–∏—Ü–∏–∞–ª—ã, –æ—Ç—Å–æ—Ä—Ç–∏—Ä—É–π—Ç–µ –∏—Ö –≤ –∞–ª—Ñ–∞–≤–∏—Ç–Ω–æ–º –ø–æ—Ä—è–¥–∫–µ –∏ –ø–æ–º–µ—Å—Ç–∏—Ç–µ –≤ —Å—Ç—Ä–æ–∫—É, —Ä–∞–∑–¥–µ–ª–µ–Ω–Ω—É—é –ø—Ä–æ–±–µ–ª–æ–º.
' '.join(sorted(planet_to_initial.values()))
```

–û—á–µ–Ω—å –ø–æ–ª–µ–∑–Ω—ã–π –º–µ—Ç–æ–¥ dict.items() –ø–æ–∑–≤–æ–ª—è–µ—Ç –Ω–∞–º –ø–µ—Ä–µ–±–∏—Ä–∞—Ç—å –∫–ª—é—á–∏ –∏ –∑–Ω–∞—á–µ–Ω–∏—è —Å–ª–æ–≤–∞—Ä—è –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ. (–ù–∞ –∂–∞—Ä–≥–æ–Ω–µ Python —ç–ª–µ–º–µ–Ω—Ç –æ—Ç–Ω–æ—Å–∏—Ç—Å—è –∫ –ø–∞—Ä–µ –∫–ª—é—á-–∑–Ω–∞—á–µ–Ω–∏–µ)

```python
for planet, initial in planet_to_initial.items():
    print("{} begins with \"{}\"".format(planet.rjust(10), initial))
```

–ß—Ç–æ–±—ã –æ–∑–Ω–∞–∫–æ–º–∏—Ç—å—Å—è —Å –ø–æ–ª–Ω—ã–º –ø–µ—Ä–µ—á–Ω–µ–º –º–µ—Ç–æ–¥–æ–≤, –∏—Å–ø–æ–ª—å–∑—É–µ–º—ã—Ö —Å–ª–æ–≤–∞—Ä—è–º–∏, –Ω–∞–∂–º–∏—Ç–µ –∫–Ω–æ–ø–∫—É "–í—ã–≤–æ–¥" –Ω–∏–∂–µ, —á—Ç–æ–±—ã –ø–æ–ª–Ω–æ—Å—Ç—å—é –æ–∑–Ω–∞–∫–æ–º–∏—Ç—å—Å—è —Å–æ —Å—Ç—Ä–∞–Ω–∏—Ü–µ–π —Å–ø—Ä–∞–≤–∫–∏, –∏–ª–∏ –æ–∑–Ω–∞–∫–æ–º—å—Ç–µ—Å—å —Å –æ—Ñ–∏—Ü–∏–∞–ª—å–Ω–æ–π –æ–Ω–ª–∞–π–Ω-–¥–æ–∫—É–º–µ–Ω—Ç–∞—Ü–∏–µ–π.
