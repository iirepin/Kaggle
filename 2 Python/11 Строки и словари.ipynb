# Струнные¶

# Синтаксис строк
Вы уже видели множество строк в примерах на предыдущих уроках, но просто напомню, что строки в Python могут быть определены с помощью одинарных или двойных кавычек. Функционально они эквивалентны.

```python
x = 'Pluto is a planet'
y = "Pluto is a planet"
x == y
```

Двойные кавычки удобны, если ваша строка содержит символ одинарной кавычки (например, представляющий апостроф).

Аналогично, легко создать строку, содержащую двойные кавычки, если заключить ее в одинарные кавычки:

```python
print("Pluto's a planet!")
print('My dog is named "Pluto"')
```

Если мы попытаемся поместить символ одинарной кавычки в строку, заключенную в одинарные кавычки, Python запутается:

```python
'Pluto's a planet!'
```

Мы можем исправить это, "экранируя" одинарную кавычку обратной косой чертой.

```python
'Pluto\'s a planet!'
```

В таблице ниже приведены некоторые важные варианты использования символа обратной косой черты.


Последняя последовательность, \n, представляет собой символ новой строки. Это заставляет Python начинать новую строку.

```python
hello = "hello\nworld"
print(hello)
```

Кроме того, синтаксис тройных кавычек Python для строк позволяет нам включать новые строки буквально (т.е. просто нажимая "Enter" на клавиатуре, а не используя специальную последовательность "\n"). Мы уже видели это в строках документации, которые мы используем для документирования наших функций, но мы можем использовать их везде, где захотим определить строку.

```python
triplequoted_hello = """hello
world"""
print(triplequoted_hello)
triplequoted_hello == hello
```

Функция print() автоматически добавляет символ новой строки, если мы не укажем значение для ключевого слова аргумент end, отличное от значения по умолчанию '\n':

```python
print("hello")
print("world")
print("hello", end='')
print("pluto", end='')
```

# Строки - это последовательности¶
Строки можно рассматривать как последовательности символов. Почти все, что, как мы видели, мы можем сделать со списком, мы также можем сделать со строкой.

```python
# Индексация
planet = 'Pluto'
planet[0]
```

```python
# Нарезка
planet[-3:]
```

```python
# Какой длины эта строка?
len(planet)
```

```python
# Да, мы даже можем обойти их по кругу.
[char+'! ' for char in planet]
```

Но главное, чем они отличаются от списков, - это то, что они неизменяемы. Мы не можем их изменять.

```python
planet[0] = 'B'
# planet.append тоже не работает
```

# Строковые методы
Как и list, тип str имеет множество очень полезных методов. Здесь я покажу лишь несколько примеров.

```python
# ВСЕ ЗАГЛАВНЫЕ БУКВЫ
claim = "Pluto is a planet!"
claim.upper()
```

```python
# все в нижнем регистре
claim.lower()
```

```python
# Поиск первого индекса подстроки
claim.index('plan')
```

```python
claim.startswith(planet)
```

```python
# неверно из-за отсутствия восклицательного знака
claim.endswith('planet')
```

Переключение между строками и списками: .split() и .join()¶

функция str.split() преобразует строку в список строк меньшего размера, по умолчанию разбивая ее на пробелы. Это очень удобно для перехода от одной большой строки к списку слов.

```python
words = claim.split()
words
```

Иногда вам захочется выделить что-то другое, кроме пробелов:

```python
datestr = '1956-01-31'
year, month, day = datestr.split('-')
```

str.join() ведет нас в другом направлении, объединяя список строк в одну длинную строку, используя строку, для которой она была вызвана, в качестве разделителя.

```python
'/'.join([month, day, year])
```

```python
# Да, мы можем поместить символы юникода прямо в наши строковые литералы :)
' 👏 '.join([word.upper() for word in words])
```

Построение строк с помощью .format()¶

Python позволяет нам объединять строки с помощью оператора +.

```python
planet + ', we miss you.'
```

Если мы хотим добавить какие-либо нестроковые объекты, мы должны быть осторожны при вызове str() для них в первую очередь

```python
position = 9
planet + ", you'll always be the " + position + "th planet to me."
```

```python
planet + ", you'll always be the " + str(position) + "th planet to me."
```

Это становится трудно читать и раздражает, когда приходится вводить. str.format() на помощь.

```python
"{}, you'll always be the {}th planet to me.".format(planet, position)
```

Намного проще! Мы вызываем .format() для "строки формата", где значения Python, которые мы хотим вставить, представлены с помощью {}-заполнителей.

Обратите внимание, что нам даже не пришлось вызывать str() для преобразования позиции из int. функция format() позаботится об этом за нас.

Если бы это было все, что делала функция format(), она все равно была бы невероятно полезна. Но, как оказалось, она может сделать гораздо больше. Вот лишь краткий обзор:

```python
pluto_mass = 1.303 * 10**22
earth_mass = 5.9722 * 10**24
population = 52910390
#         2 decimal points   3 decimal points, format as percent     separate with commas
"{} weighs about {:.2} kilograms ({:.3%} of Earth's mass). It is home to {:,} Plutonians.".format(
    planet, pluto_mass, pluto_mass / earth_mass, population,
)
```

```python
# Ссылка на аргументы format() по индексу, начиная с 0
s = """Pluto's a {0}.
No, it's a {1}.
{0}!
{1}!""".format('planet', 'dwarf planet')
print(s)
```

Вероятно, вы могли бы написать небольшую книгу только в формате str., поэтому я остановлюсь здесь и укажу вам на pyformat.info и официальные документы для дальнейшего чтения.

# Словари¶
Словари - это встроенная структура данных Python для сопоставления ключей со значениями.

```python
numbers = {'one':1, 'two':2, 'three':3}
```

В данном случае "один", "два" и "три" являются ключами, а 1, 2 и 3 - их соответствующими значениями.

Доступ к значениям осуществляется с помощью синтаксиса квадратных скобок, аналогичного индексации в списках и строках.

```python
numbers['one']
```

Мы можем использовать тот же синтаксис для добавления другой пары ключ-значение

```python
numbers['eleven'] = 11
numbers
```

Или изменить значение, связанное с существующим ключом

```python
numbers['one'] = 'Pluto'
numbers
```

В Python есть понимание по словарю с синтаксисом, похожим на понимание по списку, которое мы видели в предыдущем руководстве.

```python
planets = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Uranus', 'Neptune']
planet_to_initial = {planet: planet[0] for planet in planets}
planet_to_initial
```

The in operator tells us whether something is a key in the dictionary

```python
'Saturn' in planet_to_initial
```

```python
'Betelgeuse' in planet_to_initial
```

Цикл for по словарю будет выполнять цикл по его ключам

```python
for k in numbers:
    print("{} = {}".format(k, numbers[k]))
```

Мы можем получить доступ к коллекции всех ключей или всех значений с помощью dict.keys() и dict.values() соответственно.

```python
# Соберите все инициалы, отсортируйте их в алфавитном порядке и поместите в строку, разделенную пробелом.
' '.join(sorted(planet_to_initial.values()))
```

Очень полезный метод dict.items() позволяет нам перебирать ключи и значения словаря одновременно. (На жаргоне Python элемент относится к паре ключ-значение)

```python
for planet, initial in planet_to_initial.items():
    print("{} begins with \"{}\"".format(planet.rjust(10), initial))
```
