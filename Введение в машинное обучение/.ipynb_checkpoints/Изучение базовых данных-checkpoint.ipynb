# Использование Pandas для ознакомления с вашими данными¶
Первым шагом в любом проекте машинного обучения является ознакомление с данными. Для этого вы будете использовать библиотеку Pandas. Pandas - это основной инструмент, который специалисты по обработке данных используют для изучения данных и манипулирования ими. Большинство людей сокращают pandas в своем коде как pd. Мы делаем это с помощью команды

```python
import pandas as pd
```

Наиболее важной частью библиотеки Pandas является фрейм данных. Фрейм данных содержит данные, которые можно представить в виде таблицы. Это похоже на таблицу в Excel или таблицу в базе данных SQL.

В Pandas есть мощные методы для большинства задач, которые вы захотите выполнить с данными такого типа.

В качестве примера мы рассмотрим данные о ценах на жилье в Мельбурне, Австралия. В практических упражнениях вы примените те же процедуры к новому набору данных, содержащему цены на жилье в Айове.

Пример данных (Мельбурн) находится по пути к файлу ../input/melbourne-housing-snapshot/melb_data.csv.

Мы загружаем и исследуем данные с помощью следующих команд:

```python
# сохраните путь к файлу в переменной для облегчения доступа
melbourne_file_path = '../input/melbourne-housing-snapshot/melb_data.csv'
# считайте данные и сохраняйте их во фрейме данных под названием melbourne_data.
melbourne_data = pd.read_csv(melbourne_file_path) 
# распечатайте сводку данных в Melbourne data
melbourne_data.describe()
```

# Интерпретация описания данных¶
Результаты показывают 8 чисел для каждого столбца в исходном наборе данных. Первое число, count, показывает, сколько строк содержат значения без пропусков.

Пропущенные значения возникают по многим причинам. Например, размер 2-й спальни не был бы получен при обследовании дома с 1 спальней. Мы еще вернемся к теме недостающих данных.

Второе значение - это среднее значение, которое является усредненным. При этом std - это стандартное отклонение, которое измеряет, насколько сильно распределены значения в числовом выражении.

Чтобы интерпретировать их в виде значений 25%, 50%, 75% и max, представьте, что вы сортируете каждый столбец от наименьшего значения к наибольшему. Первое (наименьшее) значение является основным. Если вы пройдете четверть списка, то найдете число, которое больше 25% значений и меньше 75% значений. Это значение составляет 25% (произносится как "25-й процентиль"). 50-й и 75-й процентили определяются аналогично, а максимальное значение - это наибольшее число.
