# Часть 3: Ваша первая отправка
Помните о нашей цели: мы хотим найти закономерности в train.csv, которые помогут нам предсказать, выжили ли пассажиры в test.csv.

Поначалу поиск закономерностей может показаться утомительным, когда нужно разобраться с таким количеством данных. Итак, начнем с простого.

# Исследуем закономерность
Помните, что в примере файла отправки в gender_submission.csv предполагается, что все пассажиры женского пола выжили (а все пассажиры мужского пола погибли).

Является ли это разумным первым предположением? Мы проверим, верна ли эта закономерность в данных (в train.csv).

Скопируйте приведенный ниже код в новую ячейку с кодом. Затем запустите ячейку.

```python
women = train_data.loc[train_data.Sex == 'female']["Survived"]
rate_women = sum(women)/len(women)

print("% of women who survived:", rate_women)
```

Прежде чем двигаться дальше, убедитесь, что ваш код возвращает результат, указанный выше. Приведенный выше код вычисляет процент выживших женщин-пассажирок (в train.csv).

Затем запустите приведенный ниже код в другой ячейке кода:

```python
men = train_data.loc[train_data.Sex == 'male']["Survived"]
rate_men = sum(men)/len(men)

print("% of men who survived:", rate_men)
```

Приведенный выше код рассчитывает процент выживших пассажиров мужского пола (в train.csv).

Из этого можно видеть, что выжили почти 75% женщин, находившихся на борту, в то время как только 19% мужчин смогли рассказать об этом. Поскольку пол, по-видимому, является таким важным показателем выживания, файл отправки в формате gender_submission.csv - неплохое первое предположение!

Но, в конечном счете, это гендерное исследование основывает свои прогнозы только на одной колонке. Как вы можете себе представить, рассматривая несколько колонок, мы можем обнаружить более сложные закономерности, которые потенциально могут дать более обоснованные прогнозы. Поскольку довольно сложно рассматривать несколько столбцов одновременно (или потребовалось бы много времени, чтобы рассмотреть все возможные шаблоны во многих разных столбцах одновременно), мы будем использовать машинное обучение, чтобы автоматизировать это для нас.

# Ваша первая модель машинного обучения
Мы построим так называемую модель случайного леса. Эта модель состоит из нескольких "деревьев" (на рисунке ниже три дерева, но мы построим 100!), которые будут индивидуально учитывать данные каждого пассажира и голосовать за то, выжил ли этот человек. Затем модель случайного леса принимает демократичное решение: побеждает результат, набравший наибольшее количество голосов!

Приведенная ниже ячейка с кодом выполняет поиск шаблонов в четырех разных столбцах данных ("Pclass", "Sex", "SibSp" и "Parch"). Он создает деревья в модели случайного леса на основе шаблонов в файле train.csv, прежде чем генерировать прогнозы для пассажиров в test.csv. Код также сохраняет эти новые прогнозы в файле CSV submission.csv.

Скопируйте этот код в свой блокнот и запустите его в новой ячейке кода.

```python
from sklearn.ensemble import RandomForestClassifier

y = train_data["Survived"]

features = ["Pclass", "Sex", "SibSp", "Parch"]
X = pd.get_dummies(train_data[features])
X_test = pd.get_dummies(test_data[features])

model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(X, y)
predictions = model.predict(X_test)

output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions})
output.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")
```

Убедитесь, что ваш блокнот выдает то же сообщение, что и выше (ваша заявка была успешно сохранена!), прежде чем двигаться дальше.
