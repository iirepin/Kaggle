Вы понимаете, что для единообразия все оценки должны быть отформатированы одинаково, и вы решаете отформатировать их все в виде букв. Для преобразования вы решаете присвоить:

"A" - любая оценка 90-100, включительно

Напишите функцию get_grade(), которая принимает в качестве входных данных:

оценка - целое число от 0 до 100, соответствующее числовой оценке

Оценка 85 соответствует оценке B. Другими словами, get_grade(85) должен возвращать "B".

# Вопрос 2

В упражнении для предыдущего урока вы написали функцию cost_of_project(), которая вычисляла цену колец для интернет-магазина, продающего кольца с пользовательской гравировкой. В этой функции не использовались условные выражения. В этом упражнении вы перепишете функцию, чтобы использовать условные операторы. Напомним, что в интернет-магазине действует следующая структура цен:

Базовая стоимость золотых колец с покрытием составляет 50 долларов, а за единицу с гравировкой взимается плата в размере 7 долларов.
Базовая стоимость колец из чистого золота составляет 100 долларов, при этом за каждую единицу с гравировкой взимается плата в размере 10 долларов.
Пробелы и знаки препинания учитываются как единицы с гравировкой.
Ваша функция cost_of_project() принимает два аргумента:

engraving - строка на языке Python с текстом гравюры
solid_gold - логическое значение, указывающее, является ли кольцо из чистого золота
Она должна возвращать стоимость проекта.

Функция была частично выполнена для вас, и вам необходимо заполнить пробелы, чтобы завершить выполнение функции.

```python
def cost_of_project(engraving, solid_gold):
    if solid_gold == True:
        cost = 100 + len(engraving) * 10
    else:
        cost = 50 + len(engraving) * 7
    return cost

# Check your answer
# q2.check()
```

# Вопрос 3

Вы программист в водохозяйственном агентстве. Недавно вам было поручено написать функцию get_water_bill(), которая принимает в качестве входных данных:

num_gallons = количество галлонов воды, которое клиент использовал в этом месяце. (Это всегда будет целое число без десятичной части.)
В результате должен быть получен счет за воду.

Агентство по водным ресурсам использует эту структуру ценообразования:

Например:

Тот, кто использует 10 000 галлонов воды в месяц, попадает на уровень 2 и должен оплатить счет за воду в размере 6 * 10 = 60 долларов. Другими словами, get_water_bill(10000) должен возвращать значение 60.0.
Тот, кто использует 25 000 галлонов воды в месяц, попадает в категорию 3 и должен оплатить счет за воду в размере 7 * 25 = 175 долларов. Другими словами, значение get_water_bill(25000) должно возвращать значение 175,0.
Не округляйте ваш ответ. Таким образом, ваш ответ может содержать доли цента.

```python
# # ЧТО НУЖНО СДЕЛАТЬ: Отредактируйте функцию, чтобы она возвращала правильный счет для разных значений
# num_gallons
def get_water_bill(num_gallons):
    bill = 1000
    if 0 <= num_gallons <= 8000:
        bill = num_gallons / bill * 5
    elif 8001 <= num_gallons <= 22000:
        bill = num_gallons / bill * 6
    elif 22001 <= num_gallons <= 30000:
        bill = num_gallons / bill * 7
    else:
        bill = num_gallons / bill * 10
    return bill

# Check your answer
# q3.check()
```

# Вопрос 4

Вы работаете в компании, предоставляющей услуги передачи данных. За 100 долларов в месяц ваша компания предоставляет 15 гигабайт (ГБ) данных. Затем за любые дополнительные данные взимается плата в размере 0,10 доллара за МБ (или 100 долларов за ГБ, поскольку 1000 МБ помещаются в 1 ГБ).

Используйте следующую ячейку кода, чтобы написать функцию get_phone_bill(), которая принимает в качестве входных данных:

гб = количество ГБ, использованных клиентом за месяц
Это должно вернуть клиенту общий телефонный счет.

Например:

Клиенту, который использует 10 ГБ данных в течение одного месяца, выставляется счет всего на 100 долларов, поскольку объем использования не превышает 15 ГБ. Другими словами, get_phone_bill(10) должен возвращать значение 100.
Клиент, который использует 15,1 ГБ (или 15 ГБ + 100 МБ) данных за один месяц, увеличил объем на 0,1 ГБ, поэтому он должен заплатить 100 долларов (стоимость тарифного плана) плюс 0,10 * 100 = 10 долларов США, что составляет 110 долларов США. Другими словами, функция get_phone_bill(15.1) должна возвращать значение 110.
Не округляйте свой ответ.

```python
# # ЧТО НУЖНО СДЕЛАТЬ: Отредактируйте функцию, чтобы она возвращала правильный счет для разных значений
# в ГБ
def get_phone_bill(gb):
    bill = 100
    if gb <= 15:
        bill = 100
    else:
        bill = bill + (gb - 15) * 100
    return bill

# Check your answer
# q4.check()
```

➡️ Вопрос 5

В Мексике на продуктах и напитках с высоким содержанием насыщенных жиров, транс-жиров, сахара, натрия и/или калорий появляются предупреждающие надписи, призванные помочь потребителям сделать выбор в пользу здорового питания.

Например, на коробке с печеньем, изображенной на рисунке ниже, есть две надписи (в правом верхнем углу):

EXCESO CALORÍAS (в переводе с английского - ИЗБЫТОК КАЛОРИЙ)
EXCESO AZÚCARES (в переводе с английского - ИЗБЫТОК САХАРА).

В этом вопросе вы будете работать с функцией get_labels(), которая получает информацию о питательных свойствах пищевого продукта и печатает необходимые предупреждающие надписи. Эта функция принимает несколько входных данных:

food_type = один из "твердых" или "жидких".
порция_размер = размер одной порции (если твердая, то в граммах; если жидкая, то в миллилитрах)
калорийность_на_сервировку = калории в одной порции
, насыщенный_жир_г = граммы насыщенных жиров в одной порции
, трансжир_г = граммы транс-жиров в одной порции
, натрий_мг = мг натрия в одной порции
, сахар_г = граммы сахара в одной порции.
Обратите внимание, что часть приведенного здесь кода может показаться вам незнакомой, поскольку мы не поделились подробностями того, как работают некоторые функции, такие как excess_sugar() или excess_saturated_fat(). Но на высоком уровне это функции, которые возвращают значение True, если считается, что в пище содержится избыток сахара или насыщенных жиров соответственно. Эти функции используются в функции get_labels(), и всякий раз, когда обнаруживается избыток (сахара или насыщенных жиров, а также транс-жиров, натрия или калорий), она выводит соответствующую метку.

```python
# импорт функций, необходимых для работы get_labels
from learntools.intro_to_programming.ex4q5 import excess_sugar, excess_saturated_fat, excess_trans_fat, excess_sodium, excess_calories

def get_labels(food_type, serving_size, calories_per_serving, saturated_fat_g, trans_fat_g, sodium_mg, sugars_g):
    # Печатать сообщения на основе полученных результатов
    if excess_sugar(sugars_g, calories_per_serving) == True:
        print("EXCESO AZÚCARES / EXCESS SUGAR")
    if excess_saturated_fat(saturated_fat_g, calories_per_serving) == True:
        print("EXCESO GRASAS SATURADAS / EXCESS SATURATED FAT")
    if excess_trans_fat(trans_fat_g, calories_per_serving) == True:
        print("EXCESO GRASAS TRANS / EXCESS TRANS FAT")
    if excess_sodium(calories_per_serving, sodium_mg) == True:
        print("EXCESO SODIO / EXCESS SODIUM")
    if excess_calories(food_type, calories_per_serving, serving_size) == True:
        print("EXCESO CALORÍAS / EXCESS CALORIES")
```

В следующей ячейке кода показано, как использовать функцию get_labels() для получения предупреждающих надписей, которые должны содержаться в продукте. Начнем с болонской колбасы. Вот изображение со всей информацией о питании. Обратите внимание, что для этого продукта,

тип продукта = "твердая" (потому что болонья твердая, а не жидкая)
порция_размер = 32 (размер порции - 32 грамма)
калорийность = 110 (в одной порции 110 калорий)
насыщенный_жир_г = 2,5 (на порцию приходится 2,5 грамма насыщенных жиров)
трансжирный_г = 0 (на порцию приходится 0 граммов транс-жиров)
натрий_мг = 400 (на порцию приходится 400 мг натрия)
количество сахара = 1 (в инструкции по питанию указано, что оно <1 г, но мы округлим это значение до 1 грамма на порцию, чтобы быть в безопасности)
Введя все эти значения в функцию, мы сможем напечатать предупреждающие надписи.
