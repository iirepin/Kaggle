Используйте следующую ячейку кода, чтобы создать функцию get_expected_cost, которая принимает в качестве входных данных три переменные:

кровати - количество спален (тип данных float)
, ванны - количество ванных комнат (тип данных float)
has_basement - есть ли в доме подвал (тип данных boolean)
Он должен возвращать ожидаемую стоимость дома с такими характеристиками. Предположим, что:

ожидаемая стоимость дома с 0 спальнями и 0 ванными комнатами и без подвала составляет 80000 долларов,
каждая спальня добавляет 30000 к ожидаемой стоимости,
каждая ванная комната добавляет 10000 к ожидаемой стоимости, а
подвал добавляет 40000 к ожидаемой стоимости.
Например,

ожидаемая стоимость дома с 1 спальней, 1 ванной комнатой и без подвала составляет 80000 + 30000 + 10000 = 120000. Это значение будет рассчитано с использованием get_expected_cost(1, 1, False).
