# Вопрос 1

В конкурсе "Цены на жилье - передовые методы регрессии" вам нужно использовать такую информацию, как количество спален и ванных комнат, чтобы предсказать цену дома. Вдохновленный этим конкурсом, вы напишете для этого свою собственную функцию.

В следующей ячейке кода создайте функцию get_expected_cost(), у которой есть два аргумента:

кровати - количество спален
, ванны - количество ванных комнат
Она должна возвращать ожидаемую стоимость дома с таким количеством спален и ванных комнат. Предположим, что:

ожидаемая стоимость дома с 0 спальнями и 0 ванными комнатами составляет 80000 долларов.
каждая спальня добавляет 30000 к ожидаемой стоимости
, каждая ванная комната добавляет 10000 к ожидаемой стоимости.
Например,

ожидаемая стоимость дома с 1 спальней и 1 ванной комнатой составляет 120000, а
дома с 2 спальнями и 1 ванной комнатой - 150000.

```python
# ЧТО НУЖНО СДЕЛАТЬ: Завершите выполнение функции
def get_expected_cost(beds, baths):
    value = 80000 + beds * 30000 + baths * 10000
    return value

# Check your answer 
#q1.check()
```

# Вопрос 3

Вы декоратор дома и хотели бы использовать Python для оптимизации некоторых своих работ. В частности, вы создаете инструмент, который собираетесь использовать для расчета стоимости покраски комнаты.

В качестве первого шага определите функцию get_cost(), которая принимает в качестве входных данных:

кв. фут_волос = общая площадь стен, подлежащих покраске
, кв. фут_ потолка = площадь потолка, подлежащего покраске
, кв. фут_пер_галлон = количество квадратных футов, которые можно покрыть одним галлоном краски
. cost_per_gallon = стоимость (в долларах) одного галлона краски
Это должно возместить стоимость (в долларах) нанесения одного слоя краски на все стены и потолок. Предположим, вы можете купить точное количество краски, которое вам нужно, то есть вы можете купить неполные галлоны (например, если вам нужно 7,523 галлона, вы можете купить именно это количество, вместо того чтобы покупать 8 галлонов и тратить немного краски впустую). Не округляйте свой ответ.

```python
# ЗАДАЧА: Завершить определение функции
def get_cost(sqft_walls, sqft_ceiling, sqft_per_gallon, cost_per_gallon):
    cost = (sqft_walls + sqft_ceiling) / sqft_per_gallon * cost_per_gallon 
    return cost

# Проверьте свой ответ
# q3.check()
```

Correct

Теперь предположим, что вы больше не можете покупать по несколько галлонов краски. (Например, если вам нужно 4,3 галлона для выполнения проекта, вам нужно купить 5 галлонов краски.)

В этом новом сценарии вы создадите новую функцию get_actual_cost, которая использует те же входные данные и рассчитывает стоимость вашего проекта.

Для этого вам понадобится функция math.ceil(). Мы демонстрируем использование этой функции в приведенной ниже ячейке кода. Она принимает в качестве входных данных число и округляет его до ближайшего целого числа.

Запустите следующую ячейку кода, чтобы протестировать эту функцию на себе. Не стесняйтесь изменять значение test_value и убедитесь, что math.ceil() возвращает ожидаемое число.

```python
test_value = 2.17

rounded_value = math.ceil(test_value)
print(rounded_value)
```

Используйте следующую ячейку кода, чтобы определить функцию get_actual_cost(). Для этого вам нужно будет использовать функцию math.ceil().

Отвечая на этот вопрос, обратите внимание, что вполне допустимо определить функцию, которая использует другую функцию. Например, мы можем определить функцию round_up_and_divide_by_three, которая использует функцию math.ceil:

def round_up_and_divide_by_three(число):
    new_value = math.ceil(число)
    конечное значение = new_value / 3
    возвращает конечное значение

```python
def get_actual_cost(sqft_walls, sqft_ceiling, sqft_per_gallon, cost_per_gallon):
    cost = math.ceil((sqft_walls + sqft_ceiling) / sqft_per_gallon) * cost_per_gallon
    return cost

# Check your answer
# q5.check()
```

Correct
