Конкурс Titanic - самое известное соревнование Kaggle в области науки о данных. В этом соревновании участникам предлагается создать модель машинного обучения, которая может предсказать, выжили ли пассажиры после кораблекрушения "Титаника", основываясь на такой информации, как возраст, пол, размер семьи и номер билета.

Запустите следующую ячейку кода без изменений, чтобы загрузить и просмотреть данные "Титаника".

Пока не беспокойтесь о деталях кода - конечным результатом является то, что все данные titanic были загружены в переменную с именем titanic_data. (Чтобы научиться писать этот код самостоятельно, вы можете пройти курс Python, а затем курс Pandas.)

```python
# Загрузите данные с соревнований "Титаник".
import pandas as pd
titanic_data = pd.read_csv("../input/titanic/train.csv")

# Показать первые пять строк данных
titanic_data.head()
```

Данные для каждого пассажира представлены в отдельной строке.

В следующей кодовой ячейке определяются и выводятся значения трех переменных:

итого = общее количество пассажиров, которые поднялись на борт судна
, выжили = количество пассажиров, переживших кораблекрушение
, несовершеннолетние = количество пассажиров в возрасте до 18 лет.
Запустите ячейку с кодом без изменений. (Пока не беспокойтесь о деталях расчета этих переменных. Вы можете узнать больше о том, как вычислять эти значения, из курса Pandas.)

```python
# Общее количество пассажиров
total = len(titanic_data)
print(total)

# Количество выживших пассажиров
survived = (titanic_data.Survived == 1).sum()
print(survived)

# Количество пассажиров младше 18 лет
minors = (titanic_data.Age < 18).sum()
print(minors)
```

Так,

всего = 891 (на борту "Титаника" находился 891 пассажир),
выживших = 342 (выжили 342 пассажира) и
несовершеннолетних = 113 (113 пассажирам не исполнилось 18 лет).
В приведенной ниже ячейке с кодом замените подчеркивание (____) на код для вычисления значений еще двух переменных:

значение survived_fraction должно быть равно доле пассажиров, переживших катастрофу "Титаника".
minors_fraction - это доля пассажиров, которые были несовершеннолетними (в возрасте до 18 лет).
Для каждой переменной вашим ответом должно быть число от 0 до 1.

Если вам нужна подсказка или вы хотите просмотреть решение, вы можете перейти к следующей ячейке кода и раскомментировать соответствующие строки кода (q5.hint() и q5.solution()).

```python
# ЗАДАЧА: Введите значение переменной survived_fraction
survived_fraction = survived / total

# Выведите значение переменной
print(survived_fraction)

# ЗАДАЧА: Введите значение переменной minors_fraction
minors_fraction = minors / total 

# Выведите значение переменной
print(minors_fraction)

# НЕ УДАЛЯТЬ: Проверьте свой ответ
# q5.check()
```
