# Вступление¶
На этом уроке вы узнаете, как организовать свой код с помощью функций. Функция - это блок кода, предназначенный для выполнения определенной задачи. Как вы увидите, функции позволят вам выполнять примерно одно и то же вычисление несколько раз без дублирования кода.

# Введение в функции: простой пример
Начнем с простого примера функции. Функция add_three(), представленная ниже, принимает любое число, добавляет к нему три и затем возвращает результат.

```python
# Определите функцию
def add_three(input_var):
    output_var = input_var + 3
    return output_var
```

Каждая функция состоит из двух частей: заголовка и основного текста.

# Заголовок¶
Заголовок функции определяет имя функции и ее аргументы.

Заголовок каждой функции начинается с def, который сообщает Python, что мы собираемся определить функцию.
В примере имя функции - add_three.
В примере аргументом является input_var. Аргумент - это имя переменной, которая будет использоваться в качестве входных данных для функции. Оно всегда заключено в круглые скобки, которые появляются сразу после названия функции. (Обратите внимание, что функция также может не иметь аргументов или иметь несколько аргументов. Далее в уроке вы увидите несколько примеров этого.)
Для каждой функции за круглыми скобками, заключающими аргументы функции, должно следовать двоеточие :.
# Тело
Текст функции определяет работу, которую выполняет функция.

Каждая строка кода в тексте функции должна иметь отступ ровно в четыре пробела. Вы можете сделать это, нажав пробел четыре раза или нажав кнопку "Tab" один раз на клавиатуре. (По мере того, как вы будете больше узнавать о Python, вам может потребоваться сделать отступ в коде более чем на четыре пробела, но об этом вы узнаете позже в этом курсе.)
Функция выполняет свою работу, проводя все строки с отступом сверху вниз.
В качестве входных данных она принимает аргумент, который в примере равен input_var.
Функция создает новую переменную output_var с вычислением output_var = input_var + 3.
Затем последняя строка кода, называемая оператором return, просто возвращает значение в output_var в качестве выходных данных функции.
Приведенная выше ячейка кода просто определяет функцию, но не запускает ее. Подробности о теле функции будут более понятны после следующей ячейки кода, когда мы действительно запустим функцию.

# Как запустить (или "вызвать") функцию
Когда мы запускаем функцию, это также можно назвать "вызовом" функции.

В приведенной ниже ячейке кода мы запускаем функцию с 10 в качестве входного значения. Мы определяем новую переменную new_number, которая устанавливается на выходе функции.

```python
# Запустите функцию с 10 в качестве входных данных
new_number = add_three(10)

# Убедитесь, что значение равно 13, как и ожидалось
print(new_number)
```


Более подробно,

add_three(10) - это значение, которое мы получаем в качестве выходных данных, когда вводим 10 в качестве значения для input_var и вызываем функцию add_three(). Когда функция запускается, она выполняет весь свой основной код, сверху донизу:
Сначала она вычисляет output_var = input_var + 3, что устанавливает output_var = 13.
Последняя строка кода - это оператор return, который возвращает значение output_var, равное 13.
Установив new_number = add_three(10), мы устанавливаем new_number = 13.
Примечание: Когда мы случайно ссылаемся на функцию add_three() в этом руководстве, мы используем пустые закрывающие скобки после названия функции. Это соответствует тому, как люди обычно пишут объяснения кода на Python, и пустые круглые скобки просто дают понять, что мы ссылаемся на функцию, а не на переменную или другой объект Python. Эти круглые скобки всегда должны быть пустыми, даже если у функции есть аргументы.

# Присвоение имен функциям
В приведенном выше примере название функции было выбрано для вас. При присвоении имен вашим собственным функциям следует использовать только строчные буквы, разделяя слова символами подчеркивания, а не пробелами.

Со временем название функций будет казаться естественным, и это нормально, что поначалу оно может вызывать дискомфорт. Лучший способ научиться - это просмотреть много кода на Python.

# Более сложный пример
Теперь, когда вы разобрались с основами, мы можем перейти к примеру с более длительным вычислением.

Допустим, вы помогаете другу рассчитать его еженедельную зарплату после вычета налогов.

Они облагаются налогом в размере 12% (другими словами, 12% их зарплаты уходит на налоги, а домой они забирают только 88%), и
Им платят почасово из расчета 15 долларов в час.
Приведенная ниже функция рассчитывает зарплату на основе количества отработанных часов. Функция сложнее, чем в первом примере, поскольку содержит больше строк кода и комментариев. Как и в примере выше, функция имеет один аргумент (количество часов). В теле функции мы:

Используем значение для num_hours, чтобы указать значение для новой переменной pay_pretax.
Используйте значение pay_pretax, чтобы указать значение для новой переменной pay_aftertax.
Верните значение переменной pay_aftertax.

```python
def get_pay(num_hours):
    # Оплата до уплаты налогов, исходя из получения 15 долларов в час
    pay_pretax = num_hours * 15
    # Оплата после уплаты налогов, исходя из того, что вы находитесь в 12%-ной налоговой категории
    pay_aftertax = pay_pretax * (1 - .12)
    return pay_aftertax
```

Мы вызываем эту функцию так же, как и первую функцию. В следующей ячейке кода рассчитывается зарплата, исходя из 40 часов работы. (После уплаты налогов она составляет 528 долларов).

```python
# Рассчитать заработную плату, исходя из 40 часов работы
pay_fulltime = get_pay(40)
print(pay_fulltime)
```

Чтобы быстро рассчитать заработную плату на основе другого количества отработанных часов, вам нужно ввести в функцию другое число. Например, предположим, что ваш друг работает 32 часа. (Тогда он получит 422,40 доллара).

```python
pay_parttime = get_pay(32)
print(pay_parttime)
```

Поскольку вы написали функцию, вы можете рассчитать оплату за разные часы работы без необходимости заново переписывать весь код для расчетов.

Функции могут помочь вам избежать ошибок в коде и сэкономить много времени. В общем, при написании кода вы должны стремиться писать как можно меньше, потому что каждый раз, когда вы вводите вычисления, это еще одна возможность случайно внести опечатку или ошибку.

# Переменная "scope"
Переменные, определенные в теле функции, не могут быть доступны вне функции. Например, следующая ячейка кода содержит ошибку, поскольку pay_after tax существует только внутри функции.

```python
print(pay_aftertax)
```

Вы получите ту же ошибку, если попытаетесь напечатать pay_pretax или num_hours. По этой причине, если вам нужна какая-либо информация из функции, вам нужно убедиться, что она указана в инструкции return в конце функции.

Мы называем областью действия переменной ту часть кода, где она доступна. Переменные, определенные внутри функции (например, pay_aftertax), имеют локальную область действия только для этой функции. Однако, как вы уже видели, переменные, определенные вне всех функций (например, pay_parttime), имеют глобальную область действия и могут быть доступны в любом месте.

# Функции с несколькими аргументами
На данный момент вы узнали, как определить функцию всего с одним аргументом. Чтобы определить функцию с несколькими аргументами, вам нужно всего лишь добавить дополнительные аргументы в круглые скобки в заголовке функции и разделить их запятой.

Мы делаем это с помощью приведенной ниже функции get_pay_with_more_inputs(), которая вычисляет еженедельную зарплату на основе трех аргументов:

num_hours - количество отработанных часов за неделю
, hourly_wage - почасовая заработная плата (в долларах США в час)
tax_bracket - процент от вашей зарплаты, который удерживается для уплаты налогов.

```python
def get_pay_with_more_inputs(num_hours, hourly_wage, tax_bracket):
    # Оплата до уплаты налогов
    pay_pretax = num_hours * hourly_wage
    # Оплата после уплаты налогов
    pay_aftertax = pay_pretax * (1 - tax_bracket)
    return pay_aftertax
```

Затем, чтобы вызвать функцию, вам нужно ввести по одному значению для каждого входного сигнала, опять же через запятую.

В приведенной ниже ячейке с кодом мы рассчитываем зарплату после уплаты налогов для человека, который работает 40 часов, зарабатывает 24 доллара в час и облагается налогом в размере 22%.

```python
higher_pay_aftertax = get_pay_with_more_inputs(40, 24, .22)
print(higher_pay_aftertax)
```

Следующая ячейка кода дает тот же результат, что и при запуске get_pay(40), потому что значение hourly_wage равно 15, а значение tax_bracket равно 12%, что соответствует тому, как мы разработали get_pay.

```python
same_pay_fulltime = get_pay_with_more_inputs(40, 15, .12)
print(same_pay_fulltime)
```

В зависимости от того, как мы планируем использовать эту новую функцию get_pay_with_more_inputs(), она может быть более полезной, чем исходная функция get_pay(), поскольку она обращается к большему количеству случаев. Вместо потенциально неверного определения почасовой заработной платы и налоговой ставки, новая функция позволяет пользователю указать правильные значения. Но если вы уверены, что почасовую заработную плату и налоговую ставку менять не нужно, то новая функция просто сложнее, чем это необходимо. В общем, при определении функций вам нужно учитывать, какая гибкость вам нужна, в зависимости от вашего варианта использования.

# Функции без аргументов
Обратите внимание, что можно определить функцию без аргументов, у которой нет инструкции return. Функция print_hello() в приведенной ниже ячейке кода является примером.

```python
# Определите функцию без аргументов и без возврата
def print_hello():
    print("Hello, you!")
    print("Good morning!")
    
# Вызов функции
print_hello()
```
