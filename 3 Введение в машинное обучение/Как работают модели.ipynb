# Как работают модели

Это первый шаг, если вы новичок в машинном обучении.

# Вступление
Мы начнем с обзора того, как работают модели машинного обучения и как они используются. Это может показаться простым, если вы раньше занимались статистическим моделированием или машинным обучением. Не волнуйтесь, скоро мы перейдем к созданию мощных моделей.

На этом курсе вы научитесь строить модели по следующему сценарию:

Ваш двоюродный брат заработал миллионы долларов, спекулируя недвижимостью. Он предложил вам стать деловыми партнерами из-за вашего интереса к науке о данных. Он предоставит деньги, а вы предложите модели, которые предскажут, сколько стоят различные дома.

Вы спрашиваете своего двоюродного брата, как он предсказывал стоимость недвижимости в прошлом, и он говорит, что это всего лишь интуиция. Но дальнейшие расспросы показывают, что он определил ценовые модели домов, которые он видел в прошлом, и использует эти модели для прогнозирования стоимости новых домов, которые он рассматривает.

Машинное обучение работает таким же образом. Мы начнем с модели, называемой Деревом решений. Существуют более сложные модели, которые дают более точные прогнозы. Но деревья решений легко понять, и они являются основным строительным блоком для некоторых из лучших моделей в области науки о данных.

Для простоты мы начнем с самого простого из возможных деревьев решений.

Он делит дома только на две категории. Прогнозируемая цена для любого рассматриваемого дома - это средняя цена за прошлые периоды для домов в той же категории.

Мы используем данные, чтобы решить, как разбить дома на две группы, а затем снова определить прогнозируемую цену в каждой группе. Этот этап извлечения шаблонов из данных называется подгонкой или обучением модели. Данные, используемые для подгонки модели, называются обучающими данными.

Детали того, как модель подходит (например, как разделить данные), достаточно сложны, поэтому мы оставим их на потом. После того, как модель будет адаптирована, вы можете применить ее к новым данным для прогнозирования цен на дополнительные дома.

# Улучшение дерева принятия решений
Какое из следующих двух деревьев принятия решений с большей вероятностью будет получено в результате подгонки обучающих данных по недвижимости?


Дерево решений слева (Дерево решений 1), вероятно, имеет больше смысла, поскольку отражает реальность того, что дома с большим количеством спален, как правило, продаются по более высоким ценам, чем дома с меньшим количеством спален. Самым большим недостатком этой модели является то, что она не учитывает большинство факторов, влияющих на цену дома, таких как количество ванных комнат, размер участка, местоположение и т.д.

Вы можете учесть больше факторов, используя дерево с большим количеством "расщеплений". Они называются "более глубокими" деревьями. Дерево решений, которое также учитывает общий размер участка для каждого дома, может выглядеть следующим образом:

Вы прогнозируете цену любого дома, просматривая дерево решений, всегда выбирая путь, соответствующий характеристикам этого дома. Прогнозируемая цена на дом находится в нижней части дерева. Точка внизу, в которой мы делаем прогноз, называется листом.

Разбиения и значения на листьях будут определяться данными, поэтому вам пора ознакомиться с данными, с которыми вы будете работать.
